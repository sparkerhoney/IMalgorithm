```python
from collections import deque

# 그래프 클래스 정의
class Graph():
    def __init__(self, size):
        self.SIZE = size
        self.graph = [[0 for _ in range(size)] for _ in range(size)]

G1 = None
queue = deque()  # BFS에 사용될 큐
visitedAry = []  # 방문한 정점들을 저장할 리스트

gSize = 6
G1 = Graph(gSize)  # 크기가 6인 그래프 객체 생성

# 그래프 초기화
G1.graph[0][1] = 1
G1.graph[0][2] = 1
G1.graph[1][0] = 1
G1.graph[1][3] = 1
G1.graph[2][0] = 1
G1.graph[2][3] = 1
G1.graph[3][1] = 1
G1.graph[3][2] = 1
G1.graph[3][4] = 1
G1.graph[3][5] = 1
G1.graph[4][3] = 1
G1.graph[4][5] = 1
G1.graph[5][3] = 1
G1.graph[5][4] = 1

current = 0
queue.append(current)  # 시작 정점을 큐에 추가
visitedAry.append(current)  # 시작 정점을 방문한 정점 리스트에 추가

while len(queue) != 0:
    current = queue.popleft()  # 큐에서 정점을 꺼냄

    for vertex in range(gSize):
        if G1.graph[current][vertex] == 1:
            if vertex in visitedAry:
                pass  # 이미 방문한 정점인 경우 아무 작업도 수행하지 않음
            else:
                queue.append(vertex)  # 방문하지 않은 정점인 경우 큐에 추가

    if current not in visitedAry:
        visitedAry.append(current)  # 아직 방문하지 않은 정점인 경우 방문한 정점 리스트에 추가

print('방문순서 -->', end='')
for i in visitedAry:
    print(i, end='   ')
```

## 깊이우선탐색 (DFS)

```python
class Graph() :
    def __init__ (self, size) :
        self.SIZE = size
        self.graph = [[0 for _ in range(size)] for _ in range(size)]
        
G1 = None
stack = []
visitedAry= []

gSize = 6
G1 = Graph(gSize)
G1.graph[0][1] = 1;G1.graph[0][2] = 1;
G1.graph[1][0] = 1;G1.graph[1][3] = 1;
G1.graph[2][0] = 1;G1.graph[2][3] = 1;
G1.graph[3][1] = 1;G1.graph[3][2] = 1;G1.graph[3][4] = 1;G1.graph[3][5] = 1;
G1.graph[4][3] = 1;G1.graph[4][5] = 1;
G1.graph[5][3] = 1;G1.graph[5][4] = 1;

current = 0
stack.append(current)
visitedAry.append(current)

while (len(stack) != 0):
    next = None
    for vertex in range(gSize) :
        if G1.graph[current][vertex] == 1 :
            if vertex in visitedAry : 
                pass
            else :
                next = vertex
                break
                
    if next != None : 
        current = next
        stack.append(current)
        visitedAry.append(current)
    else : 
        current = stack.pop()
        
print('방문순서 -->', end='')
for i in visitedAry :
    print(i, end='   ')
```

```python
# 그래프 클래스 정의
class Graph():
    def __init__(self, size):
        self.SIZE = size
        self.graph = [[0 for _ in range(size)] for _ in range(size)]

G1 = None
stack = []  # DFS에 사용될 스택
visitedAry = []  # 방문한 정점들을 저장할 리스트

gSize = 6
G1 = Graph(gSize)  # 크기가 6인 그래프 객체 생성

# 그래프 초기화
G1.graph[0][1] = 1
G1.graph[0][2] = 1
G1.graph[1][0] = 1
G1.graph[1][3] = 1
G1.graph[2][0] = 1
G1.graph[2][3] = 1
G1.graph[3][1] = 1
G1.graph[3][2] = 1
G1.graph[3][4] = 1
G1.graph[3][5] = 1
G1.graph[4][3] = 1
G1.graph[4][5] = 1
G1.graph[5][3] = 1
G1.graph[5][4] = 1

current = 0
stack.append(current)  # 시작 정점을 스택에 추가
visitedAry.append(current)  # 시작 정점을 방문한 정점 리스트에 추가

while len(stack) != 0:
    next = None
    for vertex in range(gSize):
        if G1.graph[current][vertex] == 1:
            if vertex in visitedAry:
                pass  # 이미 방문한 정점인 경우 아무 작업도 수행하지 않음
            else:
                next = vertex
                break

    if next is not None:
        current = next
        stack.append(current)  # 다음 정점을 스택에 추가
        visitedAry.append(current)  # 다음 정점을 방문한 정점 리스트에 추가
    else:
        current = stack.pop()  # 다음 정점이 없는 경우 스택에서 정점을 꺼냄

print('방문순서 -->', end='')
for i in visitedAry:
    print(i, end='   ')
```

```python
def knapsack(maxWeight, rowCount, weight, money, item):
    print('## 메모리제이션 배열 ##')
    array =[[0 for _ in range(maxWeight+1)] for _ in range(rowCount + 1)]
    item_array = [['' for _ in range(maxWeight + 1)] for _ in range(rowCount + 1)]

    for row in range(1, rowCount+1):
        print(row, '개-->', end = ' ')
        for col in range(1, maxWeight+1):
            if weight[row] > col:
                array[row][col] = array[row - 1][col]
                item_array[row][col] = item_array[row - 1][col]
            else:
                value1 = money[row] + array[row-1][col-weight[row]]
                item_value1 = item[row] + " " + item_array[row-1][col-weight[row]]
                value2 = array[row-1][col]
                item_value2 = item_array[row-1][col]
                if value1 > value2 :
                    array[row][col] = value1
                    item_array[row][col] = item_value1
                else :
                    array[row][col] = value2
                    item_array[row][col] = item_value2


            print('%6d' % array[row][col], end = ' ')
            print(': %6s' % item_array[row][col], end=' ')

        print()
    return array[rowCount][maxWeight]

maxWeight = 7
rowCount = 4
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]
item = [" ", "금괴", "수정", "루비", "진주"]

knapsack(maxWeight, rowCount, weight, money, item)
```

```python
def knapsack(maxWeight, rowCount, weight, money, item):
    print('## 메모리제이션 배열 ##')
    # 메모리제이션 배열 초기화
    array = [[0 for _ in range(maxWeight + 1)] for _ in range(rowCount + 1)]
    item_array = [['' for _ in range(maxWeight + 1)] for _ in range(rowCount + 1)]

    # 동적 계획법 알고리즘 수행
    for row in range(1, rowCount + 1):
        print(row, '개-->', end=' ')
        for col in range(1, maxWeight + 1):
            if weight[row] > col:
                # 현재 아이템의 무게가 현재 용량(col)보다 크면
                # 이전 행(row-1)에서 해당 용량(col)의 최적값을 가져옴
                array[row][col] = array[row - 1][col]
                item_array[row][col] = item_array[row - 1][col]
            else:
                # 현재 아이템을 포함하는 경우와 포함하지 않는 경우의 최적값 비교
                value1 = money[row] + array[row - 1][col - weight[row]]
                item_value1 = item[row] + " " + item_array[row - 1][col - weight[row]]
                value2 = array[row - 1][col]
                item_value2 = item_array[row - 1][col]
                if value1 > value2:
                    # 현재 아이템을 포함한 경우가 더 큰 경우
                    array[row][col] = value1
                    item_array[row][col] = item_value1
                else:
                    # 현재 아이템을 포함하지 않은 경우가 더 큰 경우
                    array[row][col] = value2
                    item_array[row][col] = item_value2

            # 메모리제이션 배열 출력
            print('%6d' % array[row][col], end=' ')
            print(': %6s' % item_array[row][col], end=' ')

        print()

    return array[rowCount][maxWeight]

maxWeight = 7
rowCount = 4
weight = [0, 6, 4, 3, 5]
money = [0, 13, 8, 6, 12]
item = [" ", "금괴", "수정", "루비", "진주"]

knapsack(maxWeight, rowCount, weight, money, item)
```
